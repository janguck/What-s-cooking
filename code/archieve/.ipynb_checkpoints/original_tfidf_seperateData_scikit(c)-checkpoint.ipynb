```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from common_helper import *
from original_helper import *
```

```python
train_dr = "../input/train.json"
test_dr = "../input/test.json"
traindf, testdf = data_load(train_dr, test_dr)
```

```python
x_train, y_train = data_preprocessing(traindf)
x_test, _ = data_preprocessing(testdf)
```

```python
tfidf = TfidfVectorizer(analyzer=lambda d: d.split(',')).fit(traindf['ingredients_string'])
x_train = tfidf.fit_transform(traindf['ingredients_string']).toarray()
y_train = traindf['cuisine']
x_test = tfidf.transform(testdf['ingredients_string']).toarray()
```

```python
long_x_train = []
long_y_train = []
short_x_train = []
short_y_train = []

long_x_test = []
short_x_test = []
long_id = []
short_id = []
select_number = 10

for i,j in zip(x_train,y_train):
    if i.tolist().count(0) < x_train.shape[1]- select_number:
        long_x_train.append(i.reshape(-1))
        long_y_train.append(j)
    else:
        short_x_train.append(i.reshape(-1))
        short_y_train.append(j)
        
for i,j in zip(x_test,testdf['id'].values):
    if i.tolist().count(0) < x_test.shape[1]- select_number:
        long_x_test.append(i.reshape(-1))
        long_id.append(j)
    else:
        short_x_test.append(i.reshape(-1))
        short_id.append(j)
long_x_train = np.array(long_x_train)
short_x_train = np.array(short_x_train)
long_x_test = np.array(long_x_test)
short_x_test = np.array(short_x_test)
```

```python
long_y_train_balance = pd.DataFrame({'long_y_train':long_y_train})
short_y_train_balance = pd.DataFrame({'short_y_train':short_y_train})
```

```python
set_long_train = {}

for key, value in zip(long_y_train_balance.long_y_train.value_counts().keys(),long_y_train_balance.long_y_train.value_counts().values):
    set_long_train[key] = value
    
set_short_train = {}

for key, value in zip(short_y_train_balance.short_y_train.value_counts().keys(),short_y_train_balance.short_y_train.value_counts().values):
    set_short_train[key] = value
```

```python
set_train = {"long": len(long_y_train_balance),"short":len(short_y_train_balance)}
plt.style.use(u'ggplot')
fig = pd.DataFrame(set_train, index=[0]).transpose()[0].sort_values(ascending=False, inplace=False).plot(kind='barh')
fig.invert_yaxis()
fig = fig.get_figure()
fig.tight_layout()
```

```python
plt.style.use(u'ggplot')
fig = pd.DataFrame(set_long_train, index=[0]).transpose()[0].sort_values(ascending=False, inplace=False).plot(kind='barh')
fig.invert_yaxis()
fig = fig.get_figure()
fig.tight_layout()
```

```python
plt.style.use(u'ggplot')
fig = pd.DataFrame(set_short_train, index=[0]).transpose()[0].sort_values(ascending=False, inplace=False).plot(kind='barh')
fig.invert_yaxis()
fig = fig.get_figure()
fig.tight_layout()
```

```python
colors = list(dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS).keys())
shuffle(colors)
colors_select =[colors[i] for i in list(range(1,156,8))]
```

```python
long_cuisine = pd.DataFrame({'long_cuisine':long_y_train})
short_cuisine = pd.DataFrame({'short_cuisine':short_y_train})


for type_x, type_y, type_name in [[long_x_train,long_y_train,"long"],[short_x_train,short_y_train,"short"]]:
    pca = PCA(n_components=2)
    X_r = pca.fit(type_x).transform(type_x)

    lda = LinearDiscriminantAnalysis(n_components=2)
    X_r2 = lda.fit(type_x, type_y).transform(type_x)
    
    plt.figure(figsize=(10,7))
    lw = 2
    for color, target_name in zip(colors_select, list(set(traindf['cuisine']))):
        plt.scatter(X_r[list(long_cuisine['long_cuisine'].loc[long_cuisine['long_cuisine']==target_name].keys()), 0], X_r[list(long_cuisine['long_cuisine'].\
                    loc[long_cuisine['long_cuisine']==target_name].keys()), 1], color=color, alpha=.8, lw=lw,label=target_name)    
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('PCA of Cooking dataset ({})'.format(type_name))
    plt.show()

    plt.figure(figsize=(10,7))
    lw = 2
    for color, target_name in zip(colors_select, list(set(traindf['cuisine']))):
        plt.scatter(X_r2[list(long_cuisine['long_cuisine'].loc[long_cuisine['long_cuisine']==target_name].keys()), 0], X_r2[list(long_cuisine['long_cuisine'].\
                    loc[long_cuisine['long_cuisine']==target_name].keys()), 1], color=color, alpha=.8, lw=lw,label=target_name)    
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title('LDA of Cooking dataset ({})'.format(type_name))
    plt.show()
```

```python
long_clf = RandomForestClassifier(n_estimators=150)
long_clf.fit(long_x_train,long_y_train)
short_clf = MultinomialNB(alpha=0.1)
short_clf.fit(short_x_train,short_y_train)
```

```python
check_predictions = long_clf.predict(long_x_train)
long_predictions = long_clf.predict(long_x_test)
print(accuracy_score(long_y_train, check_predictions))
check_predictions = short_clf.predict(short_x_train)
short_predictions = short_clf.predict(short_x_test)
print(accuracy_score(short_y_train, check_predictions))
```

```python
testdf['id'] = long_id + short_id
predictions = np.concatenate((long_predictions, short_predictions), axis=0)
save_submission(testdf, predictions)
```
