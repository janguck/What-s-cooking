```python
tfidf = TfidfVectorizer(analyzer=lambda d: d.split(',')).fit(x_train_word)

x_train_word = tfidf.fit_transform(x_train_word).toarray()
x_test_word = tfidf.transform(x_test_word).toarray()
```
