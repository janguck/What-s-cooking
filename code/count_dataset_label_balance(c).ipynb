```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
from collections import Counter
```

```python
train_df=pd.read_json('../input/train.json')
submission_df=pd.read_json('../input/test.json')
```

```python
train_ratio = train_df['cuisine'].value_counts() / len(train_df)
```

```python
set_train = {}

for key, value in zip(train_ratio.keys(),train_ratio.values):
    set_train[key] = value * 100
```

```python
test_set_label = list(set(train_df.cuisine))
test_set_label.sort()
```

```python
test_set_value = np.array([0.01327,0.02051,0.03771,
0.06436,0.0193,0.06948,0.02855,0.07672,0.01528,0.19267,0.01166,0.0364,
0.02061,0.16291,0.01981,0.01367,0.1083,0.02262,0.04364,0.02242])*100
```

```python
set_test = {}
for label, value in zip(test_set_label,test_set_value):
    set_test[label]= value
```

```python
set_train = Counter(set_train)
set_test = Counter(set_test)
```

```python
plt.style.use(u'ggplot')
fig = pd.DataFrame(set_train, index=[0]).transpose()[0].sort_values(ascending=False, inplace=False).plot(kind='barh')
fig.invert_yaxis()
fig = fig.get_figure()
fig.tight_layout()
```

```python
plt.style.use(u'ggplot')
fig = pd.DataFrame(set_test, index=[0]).transpose()[0].sort_values(ascending=False, inplace=False).plot(kind='barh')
fig.invert_yaxis()
fig = fig.get_figure()
fig.tight_layout()
```
