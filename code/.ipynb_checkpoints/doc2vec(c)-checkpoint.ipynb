```python
from gensim.models import doc2vec
from collections import namedtuple
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
from common_helper import *
```

```python
train_dr = "../input/train.json"
test_dr = "../input/test.json"
traindf, testdf = data_load(train_dr, test_dr)
```

```python
train_docs = [(token, label) for token, label in zip(traindf['ingredients'],traindf['cuisine'])]
```

```python
TaggedDocument = namedtuple('TaggedDocument', 'words tags')
tagged_train_docs = [TaggedDocument(d, [c]) for d, c in train_docs]
```

```python
doc_vectorizer = doc2vec.Doc2Vec(vector_size=300, alpha=0.025, min_alpha=0.025, seed=1234)
doc_vectorizer.build_vocab(tagged_train_docs)

for epoch in range(10):
    doc_vectorizer.train(tagged_train_docs,total_examples=doc_vectorizer.corpus_count,epochs=doc_vectorizer.epochs)
    doc_vectorizer.alpha -= 0.002 
    doc_vectorizer.min_alpha = doc_vectorizer.alpha
```

```python
train_vector = [doc_vectorizer.infer_vector(i) for i in traindf['ingredients']]
test_vector = [doc_vectorizer.infer_vector(i) for i in testdf['ingredients']]
```

```python
lda_visualization(train_vector,traindf)
```

```python
clf = RandomForestClassifier(n_estimators=150)
clf.fit(np.array(train_vector),traindf['cuisine'])
```

```python
check_predictions = clf.predict(train_vector)
predictions = clf.predict(test_vector)
accuracy_score(traindf['cuisine'], check_predictions)
save_submission(testdf, predictions)
```
