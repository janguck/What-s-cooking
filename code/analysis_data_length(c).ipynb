```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
```

```python
traindf = pd.read_json("../input/train.json")
testdf = pd.read_json("../input/test.json")
```

```python
traindf['ingredients_count_token'] = traindf['ingredients'].str.len()
```

```python
print("총 데이터의 평균은",round(sum(traindf['ingredients_count_token'])/len(traindf['ingredients_count_token']),2),"개 입니다.")
```

```python
plt.plot(traindf['ingredients_count_token'])
plt.title('Total case')
plt.show()
```

```python
traindf['cuisine'].loc[traindf['ingredients_count_token']]
plt.figure(figsize=(25,5))
sns.violinplot(x='cuisine', y='ingredients_count_token', data=traindf)
plt.xlabel('cuisine', fontsize=12)
plt.ylabel('ingredients_count_token', fontsize=12)
plt.show()
```

```python
total_label = list(set(traindf.cuisine))
total_label.sort()
```

```python
for i in total_label:
    plt.plot(traindf['ingredients_count_token'].loc[traindf['cuisine']== i])
    plt.title(i)
    plt.show()
```
